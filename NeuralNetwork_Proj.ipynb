**Load the data**

````
import pandas as pd
import h5py
import numpy as np

data = h5py.File('/content/drive/My Drive/Colab Notebooks/SVHN_single_grey1.h5', 'r')
````

````
print(list(data.keys()))
````

**Extract the data into variables**

````
X_train = data.get('X_train')
y_train = data.get('y_train')
X_val = data.get('X_val')
y_test = data.get('y_test')
X_test = data.get('X_test')
y_val = data.get('y_val')
````

````
X_train=np.array(X_train)
X_test=np.array(X_test)
X_val=np.array(X_val)
````

````
X_train=X_train.reshape(42000,32*32)
X_test=X_test.reshape(18000,32*32)
X_val=X_val.reshape(60000,32*32)
````

Perform KNN classifier :

````
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# Since KNN is taking too much time to complete one cycle, considering optimum K value as 5

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
print("k=%d, accuracy=%.2f%%" % (5, (model.score(X_val, y_val)) * 100))
predictions = model.predict(X_test)
````

**Report & Confusion matrix**

```
# This is formatted as code
```



````
print("Classification Report :")
print(classification_report(y_test, predictions))

print ("Confusion matrix")
print(confusion_matrix(y_test,predictions))
````

**Observations on KNN :**


1.   KNN is taking very long time to complete which is not applicable
2.   The accuracy, precision and recall are also not very commendable
3.   More than half of the classifications are incorrect for each classifier as per the confusion matrix



**Load the data into Arrays to be loaded into the Neural Network**

````
import tensorflow as tf

X_train = np.array(data.get('X_train'))
y_train = np.array(data.get('y_train'))
X_val = np.array(data.get('X_val'))
y_test = np.array(data.get('y_test'))
X_test = np.array(data.get('X_test'))
y_val = np.array(data.get('y_val'))

tf.reset_default_graph()
y_train=tf.keras.utils.to_categorical(y_train,num_classes=10)
y_val=tf.keras.utils.to_categorical(y_val,num_classes=10)
#y_test=tf.keras.utils.to_categorical(y_test,num_classes=10)
````

**Create a Keras graph with input layers, hidden layers and output**

````
model = tf.keras.models.Sequential()
````

````
model.add(tf.keras.layers.Reshape((1024,),input_shape=(32,32,)))
model.add(tf.keras.layers.BatchNormalization())
````

````
model.add(tf.keras.layers.Dense(150,activation='relu'))
model.add(tf.keras.layers.Dense(100,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='softmax'))
model.compile(optimizer='sgd',loss='categorical_crossentropy',metrics=['accuracy'])
````

````
model.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=15,batch_size=42)
````

````
predictions = model.predict_classes(X_test)
````

**Accuracy of Neural Network on Test data**

````
model.evaluate(X_test, tf.keras.utils.to_categorical(y_test,num_classes=10))
````

**Report & Confusion matrix**

````
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
````

**Observations on Neural Network :**


1.   A lot faster than KNN
2.   Accuracy, recall and precision of Neural Network are impressive
